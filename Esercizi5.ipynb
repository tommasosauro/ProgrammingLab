```python
#Esercizio 1
import numpy as np
import pandas as pd

sales = pd.DataFrame(
    data={
        "employee": [
            "Katrina",
            "Guanyu",
            "Jan",
            "Roman",
            "Jacqueline",
            "Paola",
            "Esperanza",
            "Alaina",
            "Egweyn",
        ],
        "sales": [14, 17, 6, 12, 8, 3, 7, 15, 5],
        "year": [2018, 2019, 2020, 2018, 2020, 2019, 2019, 2020, 2020],
    }
)

print("Vendite maggiori di 10:")
vendite_maggiori_10 = sales[sales['sales']>10]
print(vendite_maggiori_10)

print("Dati del 2018:")
dati_2018 = sales[sales['year']==2018]
print(dati_2018)

print("Tutto tranne maggiori di 13 nel 2018:")
tranne_vendite_2018 = sales[~((sales['sales'] > 13) & (sales['year'] == 2018))]
print(tranne_vendite_2018)

print("Vendite divisi per 3 maggiori di 3:")
diviso_3 = sales[sales['sales']/3 > 3]
print(diviso_3)

print("Dipendenti con prima lettera dopo la J:")
lettera_j = sales[sales['employee'] > 'J']
print(lettera_j)
```

```python
#Esercizio 2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)

videogiochi = df["Name"].value_counts()
print("Sono stati pubblicati", len(videogiochi), "videogiochi")
print("\n")

#Generi pi popolari
generi = df["Genre"].value_counts()
print(generi)
plt.figure(figsize=(12,6))
plt.plot(generi)
plt.title("Genere pi popolari")
plt.xlabel("Generi")
plt.ylabel("Popolarit")
plt.grid(False)
plt.show()
print("\n")

#Evoluzione videogiochi nel tempo
videogiochi_negli_anni = df["Year"].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.plot(videogiochi_negli_anni.values , videogiochi_negli_anni.index)
plt.title("Evoluzione videogiochi negli anni")
plt.xlabel("Anni")
plt.ylabel("Aumento videogiochi")
plt.grid(False)
plt.show()
print("\n")

#Riprodurre il grafico dell'esercizio

sales_per_genere = df.groupby("Genre")["Global_Sales"].sum().sort_values()

plt.figure(figsize=(12,6))
plt.bar(sales_per_genere.index, sales_per_genere.values)
plt.title("Vendite Globabili per Genere")
plt.xlabel("Generi")
plt.grid(False)
plt.show()

```
