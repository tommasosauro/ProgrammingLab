```python
#Esercizio 1

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import math

#Distribuzione
np.random.seed(42)
parametro_forma = 1
campione = stats.gamma.rvs(a=parametro_forma, size=1000)
print(campione)

#Istrogramma con PDF
x = np.linspace(0, campione.max(), 100)
pdf = stats.gamma.pdf(x,a=parametro_forma)
plt.plot(x,pdf,'r')
plt.title("PDF")

plt.hist(campione, bins=30, alpha=0.5)
plt.title("Istogramma del campione con PDF")
plt.xlabel("Campioni")
plt.ylabel("Frequenza")

#Modello Fit
parametri = stats.gamma.fit(campione, floc=0)
forma_stimata = parametri[0]
print("Parametro di forma stimato dal campione:", forma_stimata)
print("Parametro di forma originale:", parametro_forma)
```
