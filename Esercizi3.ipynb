```python
#Esercizio 1

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv")

df.sort_values("total_litres_of_pure_alcohol", ascending=False, inplace=True)
print(df[['country', 'total_litres_of_pure_alcohol']].head(10))

media_beer = df['beer_servings'].mean()
media_wine = df['wine_servings'].mean()
media_spirit = df['spirit_servings'].mean()

print(media_beer)
print(media_wine)
print(media_spirit)

df['alcohol_index'] = (df['beer_servings'] + df['wine_servings'] + df['spirit_servings'])/3

paese_maggiore = df.loc[df['alcohol_index'].idxmax()]
print(paese_maggiore)

paese_piu_100_birre = df[df['beer_servings']>100]
print(paese_piu_100_birre[['country','beer_servings']])

plt.figure(figsize=(10,5))
plt.bar(df['country'].head(10), df['total_litres_of_pure_alcohol'].head(10))
plt.title("Top 10 paesi per consumo di alcol")
plt.xlabel("Paesi")
plt.ylabel("Litri di alcol puro")
plt.show()


vini_ordinati = df.sort_values(('wine_servings'),ascending=False)
plt.figure(figsize=(10,5))
plt.plot(vini_ordinati['country'],vini_ordinati['wine_servings'],marker='o')
plt.title("Consumo di vino per paese")
plt.xlabel("Paesi")
plt.ylabel("Porzioni di vino")
plt.show()
         
```

```python
#Esercizio 2

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df.sort_values('salary_year_avg', ascending=False, inplace=True)

plt.figure(figsize=(10,5))
plt.bar(df['salary_year_avg'].head(20),df['job_title_short'].head(20))
plt.title("Stipendio annuale medio per lavoro")
plt.xlabel("Stipendio")
plt.ylabel("Lavoro")
plt.show()
```
