```python
#Esercizio 3

import numpy as np

a = np.random.random(size=(10,3))
print(a)

differenze = np.abs(a - 0.5)
indici = np.argmin(differenze,axis=1)
valori_piu_vicini = a[np.arange(10),indici]

for i in range(10):
    print(f"Riga {i}: {a[i]}")
    print(f"L'elemento più vicino a 0.5 della riga è: {valori_piu_vicini[i]}")
```

```python
#Esercizio 4

import numpy as np

frequenza_cardiaca = np.array([68, 65, 77, 110, 160, 161, 162, 161, 160, 161, 162, 163, 
                               164, 163, 162, 100, 90, 97, 72, 60, 70])

frequenza_minima = np.amin(frequenza_cardiaca)
print(frequenza_minima)

frequenza_massima = np.amax(frequenza_cardiaca)
print(frequenza_massima)

attivita_fisica = np.where(frequenza_cardiaca > 120, True, False)
print(attivita_fisica)

contatore = 0

for frequenza in attivita_fisica:
    if frequenza == True:
        contatore = contatore + 1

print(contatore/len(frequenza_cardiaca))
```

```python
#Esercizio 5

a = np.array([50000,105250,55000,89000])

a_aumento_ceo = np.where(a == 105250, a * 1.15, a)
print(a_aumento_ceo)

a_aumento_1 = np.where(a_aumento_ceo == 50000, a_aumento_ceo*1.2, a_aumento_ceo)
print(a_aumento_1)

a_aumento_2 = np.where( a_aumento_1 == a, a*1.1, a_aumento_1)
print(a_aumento_2)

totale_no_aumenti = np.sum(a)
totale_aumenti = np.sum(a_aumento_2)

print(totale_aumenti - totale_no_aumenti)
```

```python
#Esercizio 6

import numpy as np

survey_matrix = np.array([ [25, 40000, 10], [32, 52000, 12], [40, 63000, 14], [29, 47000, 11], 
                          [35, 58000, 13] ])
survey_matrix_12_anni= survey_matrix[survey_matrix[:,2] >= 12]
print("I dipendenti con almeno 12 anni di esperienza sono:\n", survey_matrix_12_anni)

survey_matrix_redditi = survey_matrix_12_anni[:,1]
print("I redditi dei dipendenti con almeno 12 anni di esperienza sono:\n", survey_matrix_redditi)

media_redditi = np.mean(survey_matrix_redditi)
print("La media dei redditi dei dipendenti con almeno 12 anni di esperienza è:", media_redditi,"dollari.")
```
