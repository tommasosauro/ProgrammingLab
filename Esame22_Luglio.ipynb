```python
#Esercizio 1

import numpy as np

vettore_spese = np.array([3200, 2750, 4100, 3600, 2950, 5100, 4400])

spesa_totale = vettore_spese.sum()
print("La spesa totale sostenuta dal gruppo è:", spesa_totale)

spesa_detratta = np.where(vettore_spese < 3000, vettore_spese*0.85, vettore_spese)
print("Le spese dopo la detrazione sono uguali a:", spesa_detratta)

spesa_finale = spesa_totale.sum() - spesa_detratta.sum()
print("La spesa finale sostenuta dal gruppo è:", spesa_finale)
```

```python
#Esercizio 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def lancio_dado(n):

    vettore_possibilita = [1,2,3,4,5,6]
    vettore_probabilita = [1/6,1/6,1/6,1/6,1/6,1/6]

    vettore_risultati = np.random.choice(vettore_possibilita, n, vettore_probabilita)
    vettore_uscite = [0,0,0,0,0,0]

    for risultato in vettore_risultati:
        vettore_uscite[risultato-1]=vettore_uscite[risultato-1]+1
    return vettore_uscite

vettore_dimensione = np.linspace(10,5000,100)
vettore_uscite = []
vettore_frequenze = []
data = []
df = pd.DataFrame(data, columns = ['1' , '2', '3', '4', '5', '6'])

for dimensione in vettore_dimensione:
    vettore_uscite = lancio_dado(int(dimensione))
    vettore_frequenze = vettore_uscite/dimensione
    data.append(vettore_frequenze)

df = pd.DataFrame(data, columns = ['1' , '2', '3', '4', '5', '6'])
print("Data Frame Frequenze")
print(df.head(100))

df = df.drop(columns='2', axis=1)
df = df.drop(columns='4', axis=1)
df = df.drop(columns='5', axis=1)
print("Rimuovo le colonne non interessate")
print(df.head(100))

plt.figure(figsize=(10,5))
plt.plot(vettore_dimensione, df['1'],label='Numero 1')
plt.plot(vettore_dimensione, df['3'],color='red',label='Numero 3')
plt.plot(vettore_dimensione, df['6'],color='green',label='Numero 6')
plt.title("Andamento delle frequenza 1")
plt.xlabel("Lanci")
plt.ylabel("Frequenza")
plt.legend()
plt.show()
```

```python
#Esercizio 3

import numpy as np
from scipy import stats

np.random.seed(0)

normale_distribuzione = stats.norm(loc=0, scale=5)
valori = normale_distribuzione.rvs(size=49)
matrice = valori.reshape(7,7)
print(matrice)

matrice_assoluta = np.abs(matrice)
indici_massimi = np.argmax(matrice_assoluta, axis=1)
valori_piu_lontani = matrice[np.arange(7), indici_massimi]

temp=1
for elemento in indici_massimi:
        print("L'elemento della", temp, "riga più lontano da zero è:", valori_piu_lontani[elemento])
        temp = temp + 1

temp2=0
for elemento in valori_piu_lontani:
    if elemento > 0:
        temp2=temp2 + 1

print("Le righe con valore più lontano da zero positive sono:", temp2)
        
```

```python
#Esercizio 4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/lauranenzi/ProgrammingLab_II/refs/heads/main/esami/data/cereal.csv'
df = pd.read_csv(url)

#Aggiungo colonna nutri_score
df['nutri_score'] = (df['protein'] + df['fiber']) / df['calories']

#Top 5 nutri_score
df = df.sort_values(df['nutri_score'], ascending=True)
cinque_cereali = df[['name','nutri_score']].head(5)
print(cinque_cereali)

#Filtro < 120 calorie e > 2 gr di fibre
df2 = df[(df['calories']<120) & (df['fiber'] > 2)]
print(df2.head())

#Grafico 10 cereali nutri score

top_10 = df.sorted.head(10)
plt.figure(figsize=(10,5))
plt.barh(top_10,['name'],top_10['nutri_score'])
plt.title("10 Cereali con il migliore nutri_score")
plt.xlabel("Cereali")
plt.ylabel("Valori")
plt.show()

#Scatterplot

plt.figure(figsize=(10,5))
plt.scatter(df['calories'],df['nutri_score'], c=df['protein'])
plt.title('Relazione tra Calories, Nutri Score e Protein')
plt.xlabel('Calories')
plt.ylabel('Nutri Score')
plt.show()
```
